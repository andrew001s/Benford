Calculate MSE

```python
def calculate_mse(frecuence, benford):
    return np.mean([(frecuence[i]-benford[i])**2 for i in range(9)])
```

```python
mse_death = calculate_mse(frecuence_death, benford)
mse_kill = calculate_mse(frecuence_kill, benford)
mse_assist = calculate_mse(frecuence_assist, benford)
mse_cs = calculate_mse(frecuence_cs, benford)
mse_gold_spent = calculate_mse(frecuence_gold_spent, benford)
mse_gold_earned = calculate_mse(frecuence_gold_earned, benford)
mse_xp = calculate_mse(frecuence_xp, benford)
mse_level = calculate_mse(frecuence_level, benford)
mse_damage_dealt = calculate_mse(frecuence_damage_dealt, benford)
mse_time_spent_dead = calculate_mse(frecuence_time_spent_dead, benford)
mse_items = calculate_mse(frecuence_items, benford)
mse_commands_pings = calculate_mse(frecuence_commands_pings, benford)
mse_damage_dealt_to_objectives = calculate_mse(frecuence_damage_dealt_to_objectives, benford)
mse_damage_dealt_to_turrets = calculate_mse(frecuence_damage_dealt_to_turrets, benford)
mse_damage_dealt_to_buildings = calculate_mse(frecuence_damage_dealt_to_buildings, benford)
mse_inhibitors_kills = calculate_mse(frecuence_inhibitors_kills, benford)
mse_items_purchased = calculate_mse(frecuence_items_purchased, benford)
mse_largest_killing_spree = calculate_mse(frecuence_largest_killing_spree, benford)
mse_time_spent_living = calculate_mse(frecuence_time_spent_living, benford)
mse_time_played = calculate_mse(frecuence_time_played, benford)
mse_total_damage_taken = calculate_mse(frecuence_total_damage_taken, benford)
mse_wards_placed = calculate_mse(frecuence_wards_placed, benford)
mse_wards_killed = calculate_mse(frecuence_wards_killed, benford)
total_mse = (mse_death+mse_kill+mse_assist+mse_cs+mse_gold_spent+mse_gold_earned+mse_xp+mse_level+mse_damage_dealt+mse_time_spent_dead+mse_items+mse_commands_pings+mse_damage_dealt_to_objectives+mse_damage_dealt_to_turrets+mse_damage_dealt_to_buildings+mse_inhibitors_kills+mse_items_purchased+mse_largest_killing_spree+mse_time_spent_living+mse_time_played+mse_total_damage_taken+mse_wards_placed+mse_wards_killed)/22
```
